```python
# Test de entorno para proyectos con LangChain en VS Code

Este notebook es un **smoke test** para comprobar que todo esta bien configurado antes de empezar mini proyectos personales.

## Que valida

1. Python y entorno activo.
2. Librerias clave para tu flujo (pandas, langchain y otras opcionales).
3. Estructura del workspace en `notebooks_anaconda`.
4. Prueba minima de una cadena simple con componentes de LangChain.

Si todo sale en verde, el entorno esta listo para trabajar conmigo desde VS Code.
```

```python
import os
import platform
import sys
from datetime import datetime
from importlib import metadata
from pathlib import Path

RESULTADOS = []


def check(nombre, condicion, detalle_ok="", detalle_fail=""):
    estado = bool(condicion)
    RESULTADOS.append((nombre, estado, detalle_ok if estado else detalle_fail))
    icono = "✅" if estado else "❌"
    detalle = detalle_ok if estado else detalle_fail
    print(f"{icono} {nombre}: {detalle}")


def detectar_raiz_workspace(actual: Path) -> Path | None:
    # Busca hacia arriba una carpeta que contenga notebooks_anaconda.
    for candidato in [actual, *actual.parents]:
        if (candidato / "notebooks_anaconda").exists():
            return candidato
    # Si estamos dentro de notebooks_anaconda, la raiz es el padre.
    if actual.name == "notebooks_anaconda" and actual.parent.exists():
        return actual.parent
    return None


print("=" * 70)
print("SMOKE TEST: VS CODE + PYTHON + LANGCHAIN")
print("=" * 70)
print(f"Fecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Sistema: {platform.system()} {platform.release()} ({platform.machine()})")
print(f"Python: {sys.version.split()[0]}")
print(f"Interprete: {sys.executable}")
print(f"Directorio actual: {os.getcwd()}")
print()

# 1) Version de Python recomendada
python_ok = sys.version_info >= (3, 10)
check(
    "Version de Python (>=3.10)",
    python_ok,
    detalle_ok=f"Correcta ({sys.version.split()[0]})",
    detalle_fail=f"Detectada {sys.version.split()[0]} - recomendado >= 3.10",
)

# 2) Paquetes clave y opcionales
paquetes = [
    "pandas",
    "langchain",
    "langchain-core",
    "langgraph",
    "openai",
]

print("\nPaquetes detectados:")
for paquete in paquetes:
    try:
        version = metadata.version(paquete)
        es_clave = paquete in {"pandas", "langchain"}
        check(
            f"Paquete {paquete}",
            True if es_clave else True,
            detalle_ok=f"instalado (v{version})",
            detalle_fail="",
        )
    except metadata.PackageNotFoundError:
        es_clave = paquete in {"pandas", "langchain"}
        check(
            f"Paquete {paquete}",
            not es_clave,
            detalle_ok="opcional, no instalado",
            detalle_fail="falta (recomendado para tus practicas)",
        )

# 3) Estructura minima del workspace
cwd = Path.cwd()
workspace_root = detectar_raiz_workspace(cwd)
ruta_notebooks = workspace_root / "notebooks_anaconda" if workspace_root else cwd / "notebooks_anaconda"
subcarpetas_esperadas = [
    "AI_Agents_in_LangGraph",
    "LangChain_LLM_App_Dev",
    "Long_Term_Agentic_Memory",
]

print("\nComprobando estructura del proyecto:")
check(
    "Raiz de proyecto detectada",
    workspace_root is not None,
    detalle_ok=str(workspace_root),
    detalle_fail="No se pudo detectar automaticamente la raiz del proyecto",
)

check(
    "Existe carpeta notebooks_anaconda",
    ruta_notebooks.exists(),
    detalle_ok=str(ruta_notebooks),
    detalle_fail="No encontrada desde la raiz detectada",
)

for sub in subcarpetas_esperadas:
    p = ruta_notebooks / sub
    check(
        f"Subcarpeta {sub}",
        p.exists(),
        detalle_ok="ok",
        detalle_fail="no encontrada",
    )

# 4) Prueba minima de LangChain (sin API key)
print("\nPrueba minima LangChain (sin modelo externo):")
try:
    from langchain_core.prompts import PromptTemplate
    from langchain_core.runnables import RunnableLambda

    prompt = PromptTemplate.from_template("Tema: {tema}")
    chain = prompt | RunnableLambda(lambda x: f"Cadena OK -> {x.to_string()}")
    salida = chain.invoke({"tema": "testing en VS Code"})

    check(
        "Ejecucion basica de cadena LangChain",
        "Cadena OK" in salida,
        detalle_ok=salida,
        detalle_fail="No se obtuvo salida valida",
    )
except Exception as e:
    check(
        "Ejecucion basica de cadena LangChain",
        False,
        detalle_ok="",
        detalle_fail=str(e),
    )

# Resumen final
print("\n" + "=" * 70)
ok = sum(1 for _, estado, _ in RESULTADOS if estado)
total = len(RESULTADOS)
print(f"RESULTADO: {ok}/{total} checks correctos")

if ok == total:
    print("\n🎉 Entorno listo para estudiar LangChain y crear mini proyectos en VS Code.")
else:
    print("\n⚠️ Hay checks pendientes. Revisa los ❌ y corrige antes de empezar proyectos grandes.")

print("=" * 70)
```
